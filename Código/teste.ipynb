```python
colrotas=[]
colrotasNOME=[]
colrotasCOORD=[]
alternativas=[]
alternativasNOME=[]
alternativasCOORD=[]
for i in range(dados.shape[0]):
    origem = "\""+ dados['RESI_GEOCODE'].iloc[i]+"\""
    destino = "\""+ dados['TRAB_GEOCODE'].iloc[i]+"\""
    h,m=dados['HORA_CHEGADA'].iloc[i].split(":")
    h = int(h)
    m = int(m)
    horaChegada = datetime(2021,10,6,h,m)
    resultado = gmaps.directions(origem, destino , mode="transit", arrival_time =horaChegada,alternatives=True )
    alternativas = []
    alternativasNOME = []
    alternativasCOORD=[]
    try:
      for alterna in range(len(resultado)):
        rota=[]
        rotaNOME=[]
        rotaCOORD=[]
        for etapa in resultado[alterna]['legs'][0]['steps']:
            if etapa['travel_mode'] == 'TRANSIT':
                rota.append(etapa['transit_details']['line']['short_name'])
                rotaNOME.append(etapa['transit_details']['line']['name'])
                rotaCOORD.append([etapa['start_location'],etapa['end_location']])
        if rota not in alternativas:
          alternativas.append(rota)
          alternativasNOME.append(rotaNOME)
          alternativasCOORD.append(rotaCOORD)
    except IndexError:
        alternativas=[]
        alternativasNOME = []
        alternativasCOORD=[]
    colrotas.append(alternativas)
    colrotasNOME.append(alternativasNOME)
    colrotasCOORD.append(alternativasCOORD)
```

```python
coord_int_temp1=[]
coord_int_temp2=[]

coord_int = []
for linha in colrotasCOORD:
  for alterna in linha:
    for i in range(1,len(alterna)):
      x = (alterna[i][0]['lat']+alterna[i-1][1]['lat'])/2
      y = (alterna[i][0]['lng']+alterna[i-1][1]['lng'])/2
      coord_int_temp1.append([x,y])
    coord_int_temp2.append(coord_int_temp1)
    coord_int_temp1=[]
  coord_int.append(coord_int_temp2)
  coord_int_temp2=[]
```
